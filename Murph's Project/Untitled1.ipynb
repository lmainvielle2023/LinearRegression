```python
import pandas as pd

playerStats = pd.read_csv(r'players.csv')
# print(playerStats)

playerStats["mpg"] = playerStats["MP"] / playerStats["G"]
# playerStats
ps_Tm = playerStats.sort_values(by=['Tm','mpg'], ascending = False)

processed_teams = []
topFive = pd.DataFrame(columns=ps_Tm.columns)
topFive.drop(topFive.index, inplace=True)
for index, row in ps_Tm.iterrows():
    Tm = row['Tm']
    if Tm not in processed_teams and Tm != 'TOT':
        # print(Tm)
        processed_teams.append(Tm)
        Tm_df = ps_Tm[ps_Tm['Tm']==Tm].head(5)
        # print(Tm_df)
        topFive = pd.concat([topFive,Tm_df])
        # print(topFive)

teamStats= topFive.groupby('Tm')['WS/48','PER','TS%', 'BPM', 'VORP', '3PAr', 'FTr','TRB%','AST%','STL%','BLK%','TOV%' ].mean()
# teamStats

nbaScores = pd.read_csv(r'updated nba scores.csv')
# nbaScores

merged_df = nbaScores.merge(teamStats, left_on='Visitor', right_on='Tm', how='left')
merged_df = merged_df.merge(teamStats, left_on='Home', right_on='Tm', suffixes=('_1', '_2'), how='left')
# merged_df


nbaScores['PTS'].astype(int)
nbaScores['PTS.1'].astype(int)
# this could possibly work

nbaScores['point_diff'] = nbaScores['PTS'] - nbaScores['PTS.1']
teamStats = teamStats.reset_index()


results = pd.DataFrame(columns=['Tm1', 'Tm2', 'WS48_diff', 'PER_diff', 'TS_diff', 'BPM_diff', 'VORP_diff', 'point_diff','TPAr_diff', 'FTr_diff','TRB_diff','AST_diff','STL_diff','BLK_diff','TOV_diff'])
for index, row in nbaScores.iterrows():
    Tm1 = row['Visitor']
    Tm2 = row['Home'] 
    tm1_stats = teamStats.loc[teamStats['Tm'] == Tm1].iloc[:, 1:].values[0]
    tm2_stats = teamStats.loc[teamStats['Tm'] == Tm2].iloc[:, 1:].values[0]
    WS48_diff = tm1_stats[0] - tm2_stats[0]
    PER_diff = tm1_stats[1] - tm2_stats[1]
    TS_diff = tm1_stats[2] - tm2_stats[2]
    BPM_diff = tm1_stats[3] - tm2_stats[3]
    VORP_diff = tm1_stats[4] - tm2_stats[4]
    TPAr_diff = tm1_stats[5] - tm2_stats[5]
    FTr_diff = tm1_stats[6] - tm2_stats[6]
    TRB_diff = tm1_stats[7] - tm2_stats[7]
    AST_diff = tm1_stats[8] - tm2_stats[8]
    STL_diff = tm1_stats[9] - tm2_stats[9]
    BLK_diff = tm1_stats[10] - tm2_stats[10]
    TOV_diff = tm1_stats[11] - tm2_stats[11]
    point_diff = row['point_diff']
    
    results = results.append({
        'Tm1': Tm1,
        'Tm2': Tm2,
        'WS48_diff': WS48_diff,
        'PER_diff': PER_diff,
        'TS_diff': TS_diff,
        'BPM_diff': BPM_diff,
        'VORP_diff': VORP_diff,
        'TPAr_diff': TPAr_diff,
        'FTr_diff': FTr_diff,
        'TRB_diff': TRB_diff,
        'AST_diff': AST_diff,
        'STL_diff': STL_diff,
        'BLK_diff': BLK_diff,
        'TOV_diff': TOV_diff,
        'point_diff': point_diff
    }, ignore_index=True)
     
results['Winning_team'] = (results['point_diff'] > 0).astype(int)
results
```

```python
import torch
import pickle
import torch.nn as nn

# Define the Net class in the new notebook
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(12, 8)
        self.fc2 = nn.Linear(8, 1)

    def forward(self, x):
        x = torch.sigmoid(self.fc1(x))
        x = torch.sigmoid(self.fc2(x))
        return x

# Load the saved model
with open("model2.pkl", "rb") as f:
    model2 = pickle.load(f)

# Make sure to set the model to evaluation mode
model2.eval()
```
